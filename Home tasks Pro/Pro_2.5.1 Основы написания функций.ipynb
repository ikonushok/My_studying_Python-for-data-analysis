## Задание 1

**Условие:** 
Напишите функцию ```multi_mean()```, которая может принимать на вход любое количество чисел и возвращает кортеж различных средних значений (среднее арифметическое, среднее геометрическое, среднее гармоническое). В ответе укажите результат работы функции для ```multi_mean(1.74, 2, 18, 24)```.

**Формат ответа:** 
Кортеж чисел (например, ```(25.4, 8.2, 3.5)```)

**Оценка в баллах:** 3

**Решение:**

````
def multi_mean(*elements):
  a = 0
  n = 0
  b = 1
  x = 0
  for i in elements:
    a += i                # сумма всех элементов
    n += 1                # число элементов
    b *= i                # произведение всех элементов
    x += 1/i              # сумма дробей 1/элемент
  arif = a / n            # среднее арифметическое
  geom = pow(b, 1/n)      # среднее геометрическое
  garm = n / x            # среднее гармоническое

  return arif, geom, garm

arif, geom, garm = multi_mean(1.74, 2, 18, 24)
print ("Среднее арифметическое = ", arif)
print ("Среднее геометрическое = ", geom)
print ("Среднее гармоническое = ", garm)
otvet = (round(arif,1), round(geom,1), round(garm,1))
print (otvet)
````
