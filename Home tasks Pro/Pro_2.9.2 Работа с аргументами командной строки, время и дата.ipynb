**Условие:** 
Методами библиотеки ```datetime``` найти количество високосных годов с начала Unix эпохи и по настоящий момент.

**Формат ответа:** 
Число (например, ```2```)

**Оценка в баллах:** 3 балла

**Решение:**

````
import datetime
from datetime import datetime, date

now = datetime.now()
year = int(now.strftime("%Y"))
lst = []

for i in range (1970 , year+1):
  if i % 4 == 0:
    lst.append(i)
    
print (lst)
print ('Число високосных годов с начала Unix эпохи: ' , len(lst))
````

## Задание 2

**Условие:** 
Методами библиотеки ```datetime``` найдите название дня недели, в который вы родились. Посчитайте также количество секунд, которые вы провели на этой планете до ```01.01.2020```.

**Формат ответа:** кортеж (например, ```('monday',123213875)```)

**Оценка в баллах:** 3 

**Решение:**

````
import datetime
from datetime import datetime


d = datetime(1976,4,3, 16, 10)    # дата рождения
day = datetime.weekday(d) # определение номера дня недели

days = ["monday","tuesday","wednesday","thursday","friday","saturday","sunday"]   # массив из дней недели
print (days[day]) # определение название дня недели

# Подсчет секунд с момента рождения до 01.01.2020
d_start = d
d_finish = datetime(2020,1,1)
print(datetime.timestamp(d_finish) - datetime.timestamp(d_start))

otvet = (days[day] , datetime.timestamp(d_finish) - datetime.timestamp(d_start))
otvet
````

**Ответ:**
