Найдите произведение $\cos^2(\pi^{25})*\sin^2(\pi^{25})$.

**Формат ответа:** Число с плавающей запятой, округленное до десятых (например,```0.1```) 

**Оценка в баллах:** 3

**Решение:**

````
# 1. Импорт модуля или модулей целиком
import math
sum1 = math.cos(pow(math.pi,25))**2 + math.sin(pow(math.pi,25))**2
print (sum1)
 
# 2. Импорт модуля с присвоением псевдонима модулю
import math as ma
sum2 = ma.cos(pow(ma.pi,25))**2 + ma.sin(pow(ma.pi,25))**2
print (sum2)

# 3. Импорт конкретного объекта из модуля
from math import cos , sin , pi
sum3 = math.cos(pow(pi,25))**2 + math.sin(pow(pi,25))**2
print (sum3)
 
# 4. Импорт конкретного объекта из модуля с присвоением псевдонима объекту
from math import cos as co
from math import sin as si
from math import pi as p
sum4 = co(pow(p,25))**2 + si(pow(p,25))**2
print (sum4)

# 5. Импорт всех объектов из модуля
from math import *
sum5 = math.cos(pow(math.pi,25))**2 + math.sin(pow(math.pi,25))**2
print (sum5)
````
