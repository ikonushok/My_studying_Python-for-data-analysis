## Задание 1

**Условие:** 
Создайте лямбда-функцию, которая находит обратное к числу значение (например, для числа ```2``` обратным является ```1/2```). Примените ее ко всем элементам листа ```[1,2,3, ..., 1000]```. Найдите сумму полученных обратных значений.

**Формат ответа:** Число (например, ```21.2```)

**Оценка в баллах:** 3

**Решение:**

````
lst = []
for i in range (1, 1001): lst.append(i)   #создали массив от 1 до 1000

obr = lambda a: 1 / a   # задание функции нахождения обратного значения

s = 0
for i in range(len(lst)):
  s += obr(lst[i])      # суммирование обратных значений из массива lst

print (s)
````
